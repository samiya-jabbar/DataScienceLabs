```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sales_data.csv')
```

```python
data.head()
```

```python
plt.plot(data['Date of Sale'], data['Total Revenue Generated'])
plt.title('Revenue over time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.bar(data['Product Name'], data['Total Revenue Generated'])
plt.title('Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.scatter(data['Number of Units Sold'], data['Total Revenue Generated'])
plt.title('Revenue vs. Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.hist(data['revenue'], bins=20)
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Count')
plt.show()
```
